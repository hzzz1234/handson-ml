```python
X = tf.placeholder(tf.float32, shape=(None,n_inputs), name="X")
y = tf.placeholder(tf.int32, shape=(None,10), name="y")
```

```python
# 低阶API
n_inputs = 28*28
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

```python
#定义全连接层
def neuron_layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2 / np.sqrt(n_inputs)
        #tf.truncated_normal(shape, mean, stddev) :shape表示生成张量的维度，mean是均值，stddev是标准差。这个函数产生正太分布，均值和标准差自己设定。这是一个截断的产生正太分布的函数，就是说产生正太分布的值如果与均值的差值大于两倍的标准差，那就重新生成。和一般的正太分布的产生随机数据比起来，这个函数产生的随机数与均值的差距不会超过两倍的标准差，但是一般的别的函数是可能的。
        init = tf.truncated_normal((n_inputs, n_neurons), stddev=stddev)
        W = tf.Variable(init, name="weights")
        b = tf.Variable(tf.zeros([n_neurons]), name="biases")
        z = tf.matmul(X,W) + b
        if activation =="relu":
            return tf.nn.relu(z)
        else:
            return z
```

```python
with tf.name_scope("dnn"):
    hidden1 = neuron_layer(X,n_hidden1,"hidden1",activation="relu")
    hidden2 = neuron_layer(hidden1,n_hidden2,"hidden2",activation="relu")
    logits = neuron_layer(hidden2,n_outputs,"outputs")
# tf.contrib.layers.fully_connected方法有同样作用，它默认是用"relu"激活函数，
# with tf.name_scope("dnn1"):
#     hidden1_1 = tf.contrib.layers.fully_connected(X,n_hidden1,scope="hidden1_1")
#     hidden2_1 = tf.contrib.layers.fully_connected(hidden1,n_hidden2,scope="hidden2_1")
#     logits_1 = tf.contrib.layers.fully_connected(hidden2,n_outputs,scope="outputs_1",activation_fn=None)
```

```python
#定义损失
with tf.name_scope("loss"):
    #这是一个TensorFlow中经常需要用到的函数。官方文档里面有对它详细的说明，
    #传入的logits为神经网络输出层的输出，shape为[batch_size，num_classes]，
    #传入的label为一个一维的vector，长度等于batch_size，每一个值的取值区间必须是[0，num_classes)，
    #其实每一个值就是代表了batch中对应样本的类别。
    #作用:计算softmax激活函数并计算交叉熵
    #和tf.nn.sparse_softmax_cross_entropy_with_logits区别:logits和labels的shape都是[batch_size, num_classes] 
    #效果:对于稀疏向量效率更高
    #xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,logits=logits)
    xentropy = tf.nn.softmax_cross_entropy_with_logits_v2(labels=y,logits=logits)
    loss = tf.reduce_mean(xentropy,name="loss")
```

```python
#定义训练过程
learning_rate = 0.01

with tf.name_scope("train"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
    train_op = optimizer.minimize(loss)
```

```python
# 评估模型,这里只使用准确性评估
with tf.name_scope("eval"):
    #表示每个样本的预测结果的前K个最大的数里面是否含有target中的值
#     correct = tf.nn.in_top_k(logits,y,1)
    correct_prediction = tf.equal(tf.argmax(logits,1), tf.argmax(y,1))

    accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/Users/zhen.huaz/work/github/ml-demo/demo_data/MNIST_data/", one_hot=True)
```

```python
#执行过程
n_epochs = 400
batch_size = 50

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(train_op,feed_dict={X:X_batch,y:y_batch})
        acc_train = accuracy.eval(feed_dict={X:X_batch,y:y_batch})
        acc_test = accuracy.eval(feed_dict={X:mnist.test.images,y:mnist.test.labels})
        
        print(epoch,"Train accuracy:",acc_train,"Test accuracy:",acc_test)
    save_path = saver.save(sess,"/tmp/my_model_final.ckpt")
```

### Fine-Tuning Network Hyperparameters
1.网络拓扑
2.深度----（简单一般优先尝试一到2层隐藏层，对于复杂任务再往上提升抽象）
3.每层neurons数----（增加层数一般优于增加单层的neurons数）
